```python
img = Image.open('/content/Lenna_(test_image).png').convert("L")
img_arr = np.array(img)
plt.imshow(img_arr)
```

```python
mean_arr = np.mean(img_arr)
center_arr = img_arr - mean_arr
cov_mat = np.cov(center_arr,rowvar=False)
plt.imshow(img_arr)
print("Covariance Matrix:\n", cov_mat)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_mat)
print("Eigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)
```

```python
sorted_eigen = np.argsort(eigenvalues)[::-1]#for descending order'
eigenvalues = eigenvalues[sorted_eigen]
eigenvectors = eigenvectors[:, sorted_eigen]
print("Sorted Eigenvalues:\n", eigenvalues)
print("\nSorted Eigenvectors:\n", eigenvectors)
```

```python
variance_ratio = eigenvalues/np.sum(eigenvalues)
cumulative_variance = np.cumsum(variance_ratio)
plt.plot(cumulative_variance)
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance Ratio")
plt.title("Cumulative Sum of Eigenvalues")
plt.grid(True)
plt.show()
```

```python
k = 10
principal_components = eigenvectors[:, :k]
print(f"Top {k} Principal Components:\n", principal_components)
```

```python
reduced_data = np.dot ( center_arr , principal_components)
```

```python
reconstructed_data = np.dot(reduced_data , principal_components.T ) + mean_arr
reconstructed_data = np.real(reconstructed_data)
```

```python
plt.imshow(img_arr)
plt.title("Original Image")
```

```python
plt.imshow(reconstructed_data)
plt.title("Reconstructed Image")
```
